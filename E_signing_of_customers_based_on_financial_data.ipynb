```python
x = dataset.drop(columns=['entry_id','e_signed'])
y = dataset['e_signed']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25, random_state = 0)
```

```python
# feature scaling or standardization over a range of values
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
# Logistic Regression Model
from sklearn.linear_model import LogisticRegression
classifr_lr = LogisticRegression(random_state=0)
classifr_lr.fit(x_train, y_train)
```

```python
y_pred = classifr_lr.predict(x_test)
from sklearn.metrics import confusion_matrix, accuracy_score
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
# Support Vector Model (SVM)
from sklearn.svm import SVC
classifr_svc = SVC(random_state=0)
classifr_svc.fit(x_train, y_train)
```

```python
y_pred = classifr_svc.predict(x_test)
from sklearn.metrics import confusion_matrix, accuracy_score
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
# RandomForest Model
from sklearn.ensemble import RandomForestClassifier
classifr_rf = RandomForestClassifier(random_state = 0)
classifr_rf.fit(x_train, y_train)
```

```python
y_pred = classifr_rf.predict(x_test)
from sklearn.metrics import confusion_matrix, accuracy_score
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
# Applying Randomized Search to find the best parameters
# using the randomforest classifier as that has the highest acc score
from sklearn.model_selection import RandomizedSearchCV
```

```python
parameters = {
    'bootstrap': [True, False],
    'max_depth':[3, 4, 5, 6, 8, 10, 12, 15],
    'max_features': ['auto', 'sqrt'],
    'min_samples_leaf': [1, 2, 4],
    'min_samples_split': [2, 5, 10],
    'n_estimators':[100, 200, 500]
}
```

```python
random_cv = RandomizedSearchCV(estimator=classifr_rf, param_distributions=parameters, 
                              scoring='roc_auc', n_jobs=-1, cv = 5, verbose=3)
```

```python
random_cv.fit(x_train, y_train)
```

```python
random_cv.best_estimator_
```

```python
random_cv.best_params_
```

```python
random_cv.best_score_
```

```python
# Final model using Random Forest Regression
from sklearn.ensemble import RandomForestClassifier
classifr = RandomForestClassifier(bootstrap=False, ccp_alpha=0.0, class_weight=None,
                       criterion='gini', max_depth=12, max_features='sqrt',
                       max_leaf_nodes=None, max_samples=None,
                       min_impurity_decrease=0.0, min_impurity_split=None,
                       min_samples_leaf=2, min_samples_split=2,
                       min_weight_fraction_leaf=0.0, n_estimators=500,
                       n_jobs=None, oob_score=False, random_state=0, verbose=0,
                       warm_start=False)
```

```python
classifr.fit(x_train, y_train)
y_pred = classifr.predict(x_test)
from sklearn.metrics import confusion_matrix, accuracy_score
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```
